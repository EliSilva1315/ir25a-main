```python
from sklearn.datasets import fetch_20newsgroups
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
newsgroupsdocs = newsgroups.data
```

```python
import pandas as pd
newsgroupsdocs_df = pd.DataFrame(newsgroupsdocs, columns=['text'])
```

2.2 Transformar a embedding

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(newsgroupsdocs)
```

```python
newsgroupsdocs_df
```

```python
embeddings
```

```python
newsgroupsdocs_df['embedding'] = embeddings.tolist()

```

```python
newsgroupsdocs_df
```

```python
doc_embeddings = np.array(newsgroupsdocs_df['embedding'].tolist())
```

```python
#borrar la columna embedding
newsgroupsdocs_df = newsgroupsdocs_df.drop(columns=['embedding'])
```

2.3 Creo una query y hago la búsqueda

```python
query ="Russian Federation"
query_emb=model.encode(query)
print(query_emb)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities = cosine_similarity(query_emb.reshape(1, -1), doc_embeddings)
```

```python
top_indices = np.argsort(similarities[0])[::-1][:5]
```

```python
print(top_indices)
```
